```python
import numpy as np

# 1. Create a Python list
my_list = [1, 2, 3, 4, 5, 6]
print("Python List:", my_list)

# 2. Convert list to NumPy array
array = np.array(my_list, dtype=int)
print("\nNumPy Array:", array)

# 3. Check array properties
print("\nArray Type:", type(array))
print("Length:", len(array))
print("Number of Dimensions:", array.ndim)
print("Shape:", array.shape)

# 4. Reshape array (3 rows, 2 columns)
array2 = array.reshape(3, 2)
print("\nReshaped Array (3x2):")
print(array2)

# 5. Reshape using automatic column size
array3 = array.reshape(3, -1)
print("\nReshape using -1:")
print(array3)
print("Dimensions:", array3.ndim)

# 6. Create multi-dimensional array
my_list2 = [1, 2, 3, 4, 5]
my_list3 = [2, 3, 4, 5, 6]
my_list4 = [9, 7, 6, 8, 9]

mul_arr = np.array([my_list2, my_list3, my_list4])

print("\nMulti-dimensional Array:")
print(mul_arr)
print("Shape:", mul_arr.shape)

# 7. Reshape multi-dimensional array
reshape_mul = mul_arr.reshape(1, 15)
print("\nReshaped Multi-dimensional Array (1x15):")
print(reshape_mul)

# 8. Array shape example
a = np.array([[1, 2, 3], [4, 5, 6]])
print("\nOriginal Array:")
print(a)
print("Shape:", a.shape)

# 9. Change shape manually
a.shape = (3, 2)
print("\nArray after changing shape to (3,2):")
print(a)

# 10. Reshape function
b = a.reshape(3, 2)
print("\nReshape using reshape() function:")
print(b)

# 11. Range example
r = range(24)
print("\nRange Example:")
print(list(r))
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# -------------------------------
# 1. Min-Max Normalization (0 to 1)
# -------------------------------

df = pd.DataFrame({
    'Age': [25, 30, 35, 40],
    'Salary': [50000, 60000, 70000, 80000]
})

print("Original Data")
print(df)

minmax_scaler = MinMaxScaler()
normalized_data = minmax_scaler.fit_transform(df)

df_normalized = pd.DataFrame(normalized_data, columns=df.columns)

print("\nMin-Max Normalized Data")
print(df_normalized)


# -------------------------------
# 2. Z-Score Normalization
# -------------------------------

standard_scaler = StandardScaler()
standardized_data = standard_scaler.fit_transform(df)

df_standardized = pd.DataFrame(standardized_data, columns=df.columns)

print("\nZ-Score Standardized Data")
print(df_standardized)


# -------------------------------
# 3. Frequency Distribution Example
# -------------------------------

marks = [12, 15, 18, 20, 22, 25, 27, 30, 32, 35,
         37, 40, 42, 45, 47, 50, 52, 55, 57, 60]

print("\nMarks Data")
print(marks)

# Step 1: Range
maximum = max(marks)
minimum = min(marks)
range_value = maximum - minimum

print("\nMaximum:", maximum)
print("Minimum:", minimum)
print("Range:", range_value)

# Step 2: Sturges Rule
n = len(marks)
k = int(round(1 + 3.322 * np.log10(n)))

print("\nNumber of Observations:", n)
print("Number of Classes (Sturges Rule):", k)

# Step 3: Class Width
class_width = int(np.ceil(range_value / k))
print("Class Width:", class_width)

# Step 4: Create Class Intervals
bins = list(range(minimum, maximum + class_width, class_width))

freq = pd.cut(marks, bins=bins, right=False)
frequency_table = pd.value_counts(freq).sort_index()

print("\nFrequency Distribution Table")
print(frequency_table)

# Step 5: Real Limits
print("\nReal Limits")
for b in bins[:-1]:
    lower = b - 0.5
    upper = b + class_width - 0.5
    print(f"{b}-{b+class_width-1} -> {lower} to {upper}")
```
